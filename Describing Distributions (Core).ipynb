```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv('Data/Medical_Dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

## VitD_levels

```python
mean = df['VitD_levels'].mean()
std = df['VitD_levels'].std()
plus_one_std = mean + std
minus_one_std = mean - std
```

```python
ax = sns.histplot(data=df, x='VitD_levels', stat='probability', kde=True)
ax.axvline(df['VitD_levels'].mean(), color = 'red')
ax.axvline(df['VitD_levels'].median(), color = 'green')
ax.axvline(plus_one_std, color = 'k')
ax.axvline(minus_one_std, color = 'k')
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0)
ax.set_title('VitD_levels');
```

```python
kurt = stats.kurtosis(df['VitD_levels'], fisher = False)
kurt
```

- Continuous
- Normal (no skew)
-Mesokurtic
